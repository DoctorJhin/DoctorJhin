```python
# grocery receipt generator - ask the user for 3 grocery items

item1_name = input("Please enter the 1st product's name:"))
item1_price = float(input("Please enter its price:"))
item1_quant = int(input("Please enter its quantity:"))
item1_total = item1_price * item1_quant

item2_name = input("Please enter the 2nd product's name:"))
item2_price = float(input("Please enter its price:"))
item2_quant = int(input("Please enter its quantity:"))
item2_total = item2_price * item2_quant

item3_name = input("Please enter the 3rd product's name:"))
item3_price = float(input("Please enter its price:"))
item3_quant = int(input("Please enter its quantity:"))
item3_total = item3_price * item3_quant

total_receipt = item1_total + item2_total + item3_total

print("Receipt")
print(item1_name, "-", item1_quant, "-", item1_total, "EGP") 
print(item2_name, "-", item2_quant, "-", item2_total, "EGP")
print(item3_name, "-", item3_quant, "-", item3_total, "EGP")
print("Total:", total_receipt)
```

```python
# salary calculator - ask for name, number of hours & hourly wage

emp_name = input("Please enter your name:"))
no_hours = int(input("Enter your working hours:"))
hourly_wage = float(input("Enter your hourly wage:"))

salary = no_hours * hourly_wage * 30

print("Your salary sums up to:", salary, "EGP")
```

```python
# travel time estimate - ask for time in hours and minute
dist = int(input("Enter the traveled distance in KMs:"))
speed = int(input("Enter your current speed:"))

time = dist / speed
hours = int(time)
minutes = int((time - hours) * 60)

print("Estimate arrival time is", hours, "hours and", minutes, "minutes")
```
